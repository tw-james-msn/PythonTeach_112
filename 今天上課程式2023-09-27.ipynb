````python
# ABCBA å‹
n = int(input('è«‹è¼¸å…¥ä¸€å€‹æ•´æ•¸: '))

for i in range(1,n+1):
  p = ' '*(n-i)
  s = '@'*i
  t = p + s
  print(t)

for i in range(n-1,1-1,-1):
  p = ' '*(n-i)
  s = '@'*i
  t = p + s
  print(t)
````

````python
# ABCBA å‹
x = int(input('è«‹è¼¸å…¥ä¸€å€‹æ•´æ•¸n:'))
n = x - 1
for i in range(-n,n+1):
  s = n+1-abs(i)
  t = x - s
  r = ' '*t + '@'*s
  print( r )
````
