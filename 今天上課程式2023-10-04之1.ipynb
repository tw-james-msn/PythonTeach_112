# list unpack

```python
x = ['高二1','李大仁',15,'男']  # list(串列)、在其它的程式語言中稱為「陣列(array)」
print( type(x[2]) ) # type() 可以查詢「值」的類別

y=x[2]  # 透過 索引(index)，可以存取list中個別的元素，其中索引的值是從0開始
z=x[-1]
print(y)
print(z)

print('-'*20)

#unpack 將list解構
a, b, *_ = x     # *表示剩下的都給「_」
print(f'a={a}')
print(f'b={b}')
print(_)

print('-'*20)

a, *_ ,b = x   #unpack
print(f'a={a}')
print(f'b={b}')
print(_)
```

# 輸入成績，然後輸出 A、B、 C、 D、 不及格

```python
score= int(input('請輸入你的成績(0-100):'))

if score>=90:
  print('你的等級是 A等')

elif score>=80:
  print('你的等級是 B等')

elif score>=70:
  print('你的等級是 C等')

elif score>=60:
  print('你的等級是 D等')

else:
  print('你的等級是 不及格')


print('執行結束')
```

```python
score= int(input('請輸入你的成績(0-100):'))

if score>=90:
  grade="A等"
elif score>=80:
  grade="B等"
elif score>=70:
  grade="C等"
elif score>=60:
  grade="D等"
else:
  grade="不及格"

print(f'你的等級是 {grade}')
print('執行結束')
```
