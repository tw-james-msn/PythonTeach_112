```python
# 定義一個函數f()
# 它接受2個參數s與n
# 執行完後會回傳n個s所組成的字串
def f(s,n):
  ss = ''
  for i in range(1,n+1):
    ss = ss + s
  return ss #回傳ss給呼叫它的程式

#主程式部份
a = 'aaaaaaaaaa'
b = f('a', 10)

print(a)
print(b)
```

```python
###########################################
# isPrime(n)函數功能：
#  透過計算傳入的 n 的因數和
#  判斷 n 是不是質數，回傳 True 及 False
###########################################

def is_prime(n):

  # 質數：大於1的自然數…
  #      它的因數只有 1 和自己(n)
  if n <= 1:
    return False

  s = 0   # 因數和
  for i in range(1, n+1):   # 1 ~ n
    if(n % i) == 0:   # 相除餘數==0
      s = s + i

  # 若因數和等於 1 + 自己(n) 那表示 n 就是質數
  # if s == 1+n:
  #   return True
  # else:
  #   return False
  return (s==1+n)

# =============
# === 主程式 ===
# =============

num = int(input("a number: "))

if is_prime(num):
  print(num, "是質數")
else:
  print(num, "不是質數")

```
