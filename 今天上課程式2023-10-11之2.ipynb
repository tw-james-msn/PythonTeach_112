```python
###########################################
# isPrime(n)å‡½æ•¸åŠŸèƒ½ï¼š
#  é€éè¨ˆç®—å‚³å…¥çš„ n çš„å› æ•¸å’Œ
#  åˆ¤æ–· n æ˜¯ä¸æ˜¯è³ªæ•¸ï¼Œå›å‚³ True åŠ False
###########################################

def is_prime(n):

  # è³ªæ•¸ï¼šå¤§æ–¼1çš„è‡ªç„¶æ•¸â€¦
  #      å®ƒçš„å› æ•¸åªæœ‰ 1 å’Œè‡ªå·±(n)
  if n <= 1:
    return False

  s = 0   # å› æ•¸å’Œ
  for i in range(1, n+1):   # 1 ~ n
    if(n % i) == 0:   # ç›¸é™¤é¤˜æ•¸==0
      s = s + i

  # è‹¥å› æ•¸å’Œç­‰æ–¼ 1 + è‡ªå·±(n) é‚£è¡¨ç¤º n å°±æ˜¯è³ªæ•¸
  # if s == 1+n:
  #   return True
  # else:
  #   return False
  return (s==1+n)

# =============
# === ä¸»ç¨‹å¼ ===
# =============

num = int(input("a number: "))

if is_prime(num):
  print(num, "æ˜¯è³ªæ•¸")
else:
  print(num, "ä¸æ˜¯è³ªæ•¸")

```
