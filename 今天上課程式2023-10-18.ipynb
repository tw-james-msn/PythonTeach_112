```python
def add(x,y):
  return x*10 + y

# -------------------------------

print(add(2,1))
print(add(y=1,x=2))
```

```python
def add(x=10,y=5):
  return x*10 + y

# -------------------------------

print(add(2))
print(add(y=7))
```

```python
def input_int(start:int=1, end:int=10, default:int=None)->int:
  '''請使用者輸入指定範圍的整數(錯誤重輸)的函數(副程式)'''
  '''https://mypy.readthedocs.io/en/stable/cheat_sheet_py3.html'''

  while True:
    n=input(f'請輸入一個({start}~{end})的整數: ')

    try:
      f=float(n or default if default!=None else n)
    except ValueError:
      continue

    i=int(f)
    if i==f and start<=i<=end:
      break

  return i

# -----------------------------------------------

n=input_int(1)
print(f'你輸入的整數是：{n}')
```
